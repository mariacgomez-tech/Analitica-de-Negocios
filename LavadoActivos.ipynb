#**Caso De Estudio**: Detección de riesgo SARLAFT en una entidad financiera
En este documento se desarrolla y se analiza un modelo Naive Bayes para la identificación temprana de clientes con alto riesgo de incurrir en lavado de activos (SARLAFT) basandose en variables de comportamiento financiero que caracterizan a una persona activa en el sistema financiero.

De acuerdo con lo anterior, estas las variables  son:

**Edad**:  Es la edad del cliente al momento del análisis.

**Ingresos**: Ingresos recibidos por una persona como resultado de una actividad comercial.

**Egresos**: Egresos o desembolsos realizados por una persona asociados al consumo, pagos o transferencias.

**Número de tarjetas de crédito**: Número total de tarjetas de crédito activas que posee el cliente dentro del sistema financiero.

**Monto transado tarjetas**: Valor total de las transacciones realizadas por el cliente mediante tarjetas de crédito o débito.

**Porcentaje crecimiento patrimonio**: Variación porcentual del patrimonio neto de un periodo a otro.

Estas variables permiten analizar tanto la capacidad económica del cliente como la intensidad y dinámica de su comportamiento transaccional, aspectos clave en la detección de posibles señales de alerta en SARLAFT.

0. Se procede con la carga de las librerías de trabajo.

```python
import numpy as np
import pandas as pd

from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
```

1. Se cargan los datos de trabajo de la base de datos Lavado de Activos

```python
nxl='/content/2. LavadoActivos.xlsx'
XDB=pd.read_excel(nxl,sheet_name=0)
XDB.head(100)

XD=XDB[['edad','ingresos_usd','gastos_usd', 'num_tarjetas_credito', 'monto_transado_tarjetas_usd','porcentaje_crecimiento_patrimonio']]
XD.head(10)
yd=XDB['lavado_activos']
yd.head()
```

2. Implementamos modelo Naive Bayes

```python
np.set_printoptions(precision=3)
mnb=GaussianNB()
mnb.fit(XD,yd)

u=mnb.theta_
sigma=mnb.var_; sigma=np.sqrt(sigma)
print("'edad','ingresos_usd','gastos_usd', 'num_tarjetas_credito', 'monto_transado_tarjetas_usd','porcentaje_crecimiento_patrimonio'")
print(u)
print("Las desviaciones son:")
print(sigma)
print()
print("Los limites superiores de las variables son: ", u+sigma)
print()
print("Los limites inferiores de las variables son: ", u-sigma)
```

#**Análisis de Resultados**
De los resultados arrojados por el modelo, podemos observar que la categoría de clientes clasificados como alto riesgo SARLAFT presenta ingresos superiores a la categoría de clientes clasificados como bajo riesgo SARLAFT, esto sugiere que niveles elevados de ingresos pueden corresponder a flujos de dinero difíciles de justificar o inconsistentes con un posible riesgo de lavado de activos. De manera similar, los egresos también son notablemente más altos en la categoría de alto riesgo. Esto indica que estos clientes no solo reciben grandes cantidades de dinero, sino que también presentan una alta circulación dentro del sistema financiero. Este factor se vuelve aún más relevante al analizar el monto transado a través de tarjetas, el cual es muy superior en los clientes de alto riesgo en comparación con los clientes clasificados como bajo riesgo.

Las variables edad, el número de tarjetas de crédito y el porcentaje de crecimeinto del patrimonio no presentan diferencias significativas entre las categorías, lo que indica que estas características no son determinantes para la clasificación.


3. Se procede con la evaluación del modelo. Para la evaluación de este tipo de modelo se utiliza la matriz de confusión.

**Desempeño del modelo**

```python
ydp=mnb.predict(XD) #Esto es lo que el modelo aprende - ydp de pronóstico
cm=confusion_matrix(yd,ydp)
print(cm)

VN=cm[0,0]; FP=cm[0,1]; FN=cm[1,0]; VP=cm[1,1]; TDatos=len(XDB)

#1. Exactitud: Funcionamiento general del modelo (Primera métrica, como funcionó pronosticando)
Ex=(VP+VN)/(TDatos)
print("Exactitud: ",Ex)

#2. Tasa de error: % fallos del sistema
TEr=(FP+FN)/(TDatos)
print("Tasa de error: ",TEr) #Modelo que pase del 75% es muy bueno

#3. Sensibilidad: Como se  comportó con respecto a los positivos. (fila)
Se=VP/(VP+FN)
print("Sensibilidad: ",Se)

#4. Especificidad: Como se comporta pronosticando negativos
Es=VN/(VN+FP)
print("Especificidad: ",Es)

#5. Precisión: Es una versión de como se comporta el modelo frente a los positivos solamente. (columna)
Pr=VP/(VP+FP)
print("Precisión: ",Pr)

#6. Predicción Negativa: como funciona el modelo pronosticando créditos PreNegados
PrN=VN/(VN+FN)
print("Predicción Negativa: ",PrN)
```

**Análisis de resultados**

De acuerdo con los resultados obtenidos, se puede observar de manera general que el modelo alcanzó una exactitud del 94%, lo que indica un muy buen desempeño en la clasificación de los clientes entre categorías de alto y bajo riesgo SARLAFT. Este resultado evidencia que el modelo logra diferenciar adecuadamente los perfiles financieros según su nivel de riesgo.

Se destaca la precisión del 94,6%, lo que indica que cuando el modelo clasifica a un cliente como de alto riesgo, en la gran mayoría de los casos  es correcto. Igualmente, se puede observar el buen comportamiento del modelo frente a la identificación de clientes de bajo riesgo, tal como lo demuestra la especificidad del 98%. Este valor indica que el modelo reconoce correctamente a casi todos los clientes que realmente no representan un riesgo significativo.

Sin embargo, la sensibilidad del 84,3% indica que algunos clientes de alto riesgo no están siendo detectados. Pero por otro lado, la predicción negativa del 93,7% confirma que la mayoría de los clientes clasificados como de bajo riesgo realmente lo son.
